```python
import pandas as pd
import os

df = pd.read_csv("../raw_data/Sample - Superstore.csv",encoding="latin1")
```

```python
df.dtypes
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Category'].unique()
```

```python
total_sales=df['Sales'].sum()
total_profit=df['Profit'].sum()
print(f"Total sales:${total_sales:,.2f}")
print(f"Total profits:${total_profit:,.2f}")
```

```python
Profit_margin = (total_profit / total_sales) * 100
print(f"Overall Profit Margin: {Profit_margin:.2f}%")
```

```python
sales_by_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print(sales_by_category)
```

```python
profit_by_category = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print(profit_by_category)
```

```python
category_summary = df.groupby('Category').agg({'Sales': 'sum', 'Profit': 'sum'})
category_summary['Profit Margin %'] = (category_summary['Profit'] / category_summary['Sales']) * 100
print(category_summary.sort_values('Profit', ascending=False))
```

```python
furniture_data = df[df['Category'] == 'Furniture']
furniture_profit = furniture_data.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=True)
print(furniture_profit)
```
